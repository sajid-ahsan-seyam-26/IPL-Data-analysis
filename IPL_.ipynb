Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Load dataset

```python
df = pd.read_csv("deliveries.csv")
df.head()
```

Basic dataset check

```python
print("Shape of dataset:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

print("\nInfo:")
print(df.info())

print("\nMissing values:")
print(df.isnull().sum())
```

NumPy code

```python
# Convert columns into NumPy arrays
total_runs_array = df["total_runs"].to_numpy()
batsman_runs_array = df["batsman_runs"].to_numpy()
wicket_array = df["is_wicket"].to_numpy()

print("Total runs in dataset:", np.sum(total_runs_array))
print("Average total runs per ball:", np.mean(total_runs_array))
print("Maximum total runs in one ball:", np.max(total_runs_array))
print("Minimum total runs in one ball:", np.min(total_runs_array))
print("Standard deviation of total runs:", np.std(total_runs_array))

print("Total batsman runs:", np.sum(batsman_runs_array))
print("Total wickets:", np.sum(wicket_array))
```

Pandas basic analysis

```python
print("First 10 rows:")
print(df.head(10))

print("\nLast 10 rows:")
print(df.tail(10))

print("\nSummary statistics:")
print(df.describe())
```

Total runs by batting team

```python
team_runs = df.groupby("batting_team")["total_runs"].sum().sort_values(ascending=False)
print(team_runs)
```

Total wickets by bowling team

```python
team_wickets = df.groupby("bowling_team")["is_wicket"].sum().sort_values(ascending=False)
print(team_wickets)
```

Top 10 batsmen by runs

```python
top_batsmen = df.groupby("striker")["batsman_runs"].sum().sort_values(ascending=False).head(10)
print(top_batsmen)
```

Top 10 bowlers by wickets

```python
top_bowlers = df.groupby("bowler")["is_wicket"].sum().sort_values(ascending=False).head(10)
print(top_bowlers)
```

Total extra runs by bowling team

```python
extra_runs_team = df.groupby("bowling_team")["extra_runs"].sum().sort_values(ascending=False)
print(extra_runs_team)
```

Plot total runs by batting team

```python
team_runs.plot(kind="bar", figsize=(12,6))
plt.title("Total Runs by Batting Team")
plt.xlabel("Batting Team")
plt.ylabel("Total Runs")
plt.xticks(rotation=90)
plt.show()
```

Plot total wickets by bowling team

```python
team_wickets.plot(kind="bar", figsize=(12,6))
plt.title("Total Wickets by Bowling Team")
plt.xlabel("Bowling Team")
plt.ylabel("Wickets")
plt.xticks(rotation=90)
plt.show()
```

Plot top 10 batsmen

```python
top_batsmen.plot(kind="bar", figsize=(12,6))
plt.title("Top 10 Batsmen by Runs")
plt.xlabel("Batsman")
plt.ylabel("Runs")
plt.xticks(rotation=45)
plt.show()
```

Plot top 10 bowlers

```python
top_bowlers.plot(kind="bar", figsize=(12,6))
plt.title("Top 10 Bowlers by Wickets")
plt.xlabel("Bowler")
plt.ylabel("Wickets")
plt.xticks(rotation=45)
plt.show()
```

Runs scored in each over

```python
runs_per_over = df.groupby("over")["total_runs"].sum()
print(runs_per_over)

runs_per_over.plot(kind="line", figsize=(10,5), marker="o")
plt.title("Runs Scored Per Over")
plt.xlabel("Over")
plt.ylabel("Runs")
plt.grid(True)
plt.show()
```

Dismissal type count

```python
dismissal_count = df["dismissal_type"].value_counts()
print(dismissal_count)
```

Plot dismissal types

```python
dismissal_count.plot(kind="bar", figsize=(12,6))
plt.title("Dismissal Types Count")
plt.xlabel("Dismissal Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Team-wise average runs per ball

```python
avg_runs_team = df.groupby("batting_team")["total_runs"].mean().sort_values(ascending=False)
print(avg_runs_team)
```

Plot average runs per ball by team

```python
avg_runs_team.plot(kind="bar", figsize=(12,6))
plt.title("Average Runs Per Ball by Batting Team")
plt.xlabel("Batting Team")
plt.ylabel("Average Runs")
plt.xticks(rotation=90)
plt.show()
```
